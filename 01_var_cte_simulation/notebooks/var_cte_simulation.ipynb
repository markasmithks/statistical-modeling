```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SEED = 42
rng = np.random.default_rng(SEED)
```

### True Normal Background

```python
x = np.linspace(Z.min(), Z.max(), 500)

true_density = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)
```

### 10000 Samples

```python
Z = rng.standard_normal(10000)

print("Mean:", np.mean(Z))
print("Std Dev:", np.std(Z))
```

```python
PRIMARY = "#c1a31f"

plt.figure()
plt.hist(Z, bins=50, density=True, color = PRIMARY,edgecolor="black",
    alpha=0.7)
plt.plot(x, true_density)
plt.title("Histogram of Standard Normal Draws with True Density Overlay")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
cumulative_mean = np.cumsum(Z) / np.arange(1, len(Z) + 1)

plt.figure()
plt.plot(cumulative_mean)
plt.title("Convergence of Sample Mean")
plt.xlabel("Number of Samples")
plt.ylabel("Cumulative Mean")
plt.show()
```

### Lognormal

```python
# Step 1: Standard normal draws
rng = np.random.default_rng(42)
Z = rng.standard_normal(10000)
```

```python
# Step 2: Convert to lognormal
mu = 0.0
sigma = 1.0

X = np.exp(mu + sigma * Z)
```

```python
# Step 3: Plot lognormal distribution
plt.figure()
plt.hist(X, bins=50, density=True, color = PRIMARY, edgecolor = "black", alpha = 0.7)
plt.title("Lognormal Distribution (Transformed from Standard Normal)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
